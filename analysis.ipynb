```python
import pandas as pd
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
sns.set_theme(rc={'figure.figsize':(11.7,8.27)}, font_scale=1.5)
```

```python
import glob
import re

# Pattern to match the model name and eei number
# mipe_5_(MODELNAME)_temp_.*?_eei_(NUMBER)_
pattern = re.compile(r'/(.*?)_temp.*?mipe_5_(.*?)_temp_.*?_eei_(\d+)_')

all_dfs = []

for filename in glob.glob('experiments/*nexus_relational_adversarial.py*.jsonl'):
    # Extract model name and eei number from the filename
    match = pattern.search(filename)
    if not match:
        # If the pattern doesn't match, skip this file
        continue
    
    model = match.group(1)
    explore_model = match.group(2)
    num_explore_iter = int(match.group(3))
    
    # Read the json lines
    df = pd.read_json(filename)
    
    # Add the new columns
    df['num_explore_iter'] = num_explore_iter
    df['explore_model'] = explore_model
    df['model'] = model
    
    all_dfs.append(df)

# Concatenate all the DataFrames
df = pd.concat(all_dfs, ignore_index=True)
df
dp = df.groupby(['model', 'explore_model', 'num_explore_iter']).agg({'success': 'mean', 'num_iter':'mean'}).reset_index()

dp['success'] = dp.success + np.random.normal(0, 0.01, size=len(dp))

dp = dp.replace({'claude-3-5-sonnet-20240620': 'Claude-3.5-Sonnet', 'claude-3-5-haiku-latest': 'Claude-3.5-Haiku', "firefunction-v2": "Firefunction-V2", "gpt-4o": "GPT-4o", "gpt-4o-mini": "GPT-4o-mini", "llama-v3p3-70b-instruct": "Llama-3.3-70B-Instruct"})
sns.lineplot(dp[dp.model == dp.explore_model], x='num_explore_iter', y='success', hue='model', marker='o')
plt.xlabel("Number of Exploration Iterations")
plt.ylabel("Success Accuracy")
plt.title("LangchainRelational-Adversarial: Task Success vs. Number of Exploration Steps")
```
